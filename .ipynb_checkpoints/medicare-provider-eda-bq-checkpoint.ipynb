```python
def cost_estimator(bq, query):
    job_config = bigquery.QueryJobConfig()
    job_config.dry_run = True
    job_config.use_query_cache = False
    query_job = bq.query(query,
        location='US',
        job_config=job_config)

    # Run query (dry run query will complete immediately)
    assert query_job.state == 'DONE'
    assert query_job.dry_run

    bytes_to_run = query_job.total_bytes_processed
    kbs_to_run = round(bytes_to_run / 1e3,2)
    mbs_to_run = round(bytes_to_run / 1e6,2)
    gbs_to_run = round(bytes_to_run / 1e9,2)

    if (len(str(gbs_to_run))) >= 1 & (gbs_to_run > 0):
        print("This query will process {} GBs".format(gbs_to_run))
    elif (len(str(mbs_to_run)) >= 1) & (mbs_to_run > 0):
        print("This query will process {} MBs".format(mbs_to_run))
    elif (len(str(kbs_to_run)) >= 1) & (kbs_to_run > 0):
        print("This query will process {} KBs".format(kbs_to_run))
    else:
        print("This query will process {} Bytes".format(bytes_to_run))
```

```python
#query = "select * from `bigquery-public-data.cms_medicare.part_d_prescriber_2014`"
#query = "select * from `bigquery-public-data.cms_medicare.referring_durable_medical_equip_2014`"
query = """select 
                npi,
                specialty_description,
                case when bene_count is null then 1 else 0 end as missing_bene_indicator, 
                sum(total_claim_count) as total_claim_count,
                sum(coalesce(bene_count,0)) as total_bene_count,
                sum(total_drug_cost) as total_drug_cost,
                sum(total_claim_count_ge65) as total_claim_count_ge65,
                sum(coalesce(bene_count_ge65,0)) as total_bene_count_ge65,
                sum(total_drug_cost_ge65) as total_drug_cost_ge65
            from `bigquery-public-data.cms_medicare.part_d_prescriber_2014`
            group by npi, specialty_description,(case when bene_count is null then 1 else 0 end)
     """
# handles org classification (only 7 orgs on this list, need to do more on this)
# flag principals somehow?
query = """
        select npi_info.*, org.npi as npi_org, case when org.npi is not null then 1 else 0 end as org_indicator
        from
        (select 
            npi,
            specialty_description,
            case when bene_count is null then 1 else 0 end as missing_bene_indicator, 
            sum(total_claim_count) as total_claim_count,
            sum(coalesce(bene_count,0)) as total_bene_count,
            sum(total_drug_cost) as total_drug_cost,
            sum(total_claim_count_ge65) as total_claim_count_ge65,
            sum(coalesce(bene_count_ge65,0)) as total_bene_count_ge65,
            sum(total_drug_cost_ge65) as total_drug_cost_ge65
        from `bigquery-public-data.cms_medicare.part_d_prescriber_2014`
        group by npi, specialty_description,(case when bene_count is null then 1 else 0 end)) npi_info
        left join (select distinct npi from `bigquery-public-data.nppes.npi_optimized` where entity_type_code_name = 'Organization') org
        on trim(npi_info.npi) = trim(org.npi)
        order by total_claim_count desc;
        """
```

```python
def grouping_helper(x, keep_list):
    if x in keep_list:
        x = x #re.sub('[^A-Za-z0-9]+','',x).lower
    else:
        x = 'Other'
    return x
```

```python
## we need to weight this by claim instead of the way we're doing it
keep_levels_specialty_description = df['specialty_description'].value_counts()[df['specialty_description'].value_counts().values > 10000].index
print(keep_levels_specialty_description)
df['specialty_desc_grouped'] = [grouping_helper(r, keep_levels_specialty_description) for r in df['specialty_description']]
df['specialty_desc_grouped'].value_counts()
```

```python
#query = "select * from `bigquery-public-data.cms_medicare.part_d_prescriber_2014`"
#query = "select * from `bigquery-public-data.cms_medicare.referring_durable_medical_equip_2014`"
query_drug = """
        select npi_info.*, org.npi as npi_org, case when org.npi is not null then 1 else 0 end as org_indicator
        from
        (select 
            npi,
            generic_name,
            specialty_description,
            case when bene_count is null then 1 else 0 end as missing_bene_indicator, 
            sum(total_claim_count) as total_claim_count,
            sum(coalesce(bene_count,0)) as total_bene_count,
            sum(total_drug_cost) as total_drug_cost,
            sum(total_claim_count_ge65) as total_claim_count_ge65,
            sum(coalesce(bene_count_ge65,0)) as total_bene_count_ge65,
            sum(total_drug_cost_ge65) as total_drug_cost_ge65
        from `bigquery-public-data.cms_medicare.part_d_prescriber_2014`
        group by npi, generic_name,specialty_description,(case when bene_count is null then 1 else 0 end)) npi_info
        left join (select distinct npi from `bigquery-public-data.nppes.npi_optimized` where entity_type_code_name = 'Organization') org
        on trim(npi_info.npi) = trim(org.npi)
        order by total_claim_count desc;
        """

query_drug = """
select 
  npi, 
  generic_name, 
  sum(total_claim_count)/sum(total_bene_count) as avgs_claim_per_beneficiary, 
  sum(total_drug_cost)/sum(total_bene_count) as avgs_cost_per_beneficiary,
  sum(total_claim_count) as total_claim_count,
  sum(total_drug_cost) as total_drug_cost,
  sum(total_bene_count) as total_bene_count
from `medicare-fraud-wids-i.drug_info.drug_info`
where specialty_description = 'Obstetrics/Gynecology'
and missing_bene_indicator = 0
group by npi, generic_name
"""
cost_estimator(bq, query_drug)
```

```python
%%time
import networkx as nx
G = nx.Graph()
G.add_weighted_edges_from(list(zip(oxy['npi'],oxy['generic_name'], oxy['avgs_claim_per_beneficiary'])))
```

```python
%%time
nx.draw(G)
```
